```python
original_signal = [15, 20, 25, 25, 30]

# Delta Encoding
delta_encoded = [original_signal[0]]  
for i in range(1, len(original_signal)):
    delta_encoded.append(original_signal[i] - original_signal[i - 1])

# Delta Decoding
decoded_signal = [delta_encoded[0]]
for i in range(1, len(delta_encoded)):
    decoded_signal.append(decoded_signal[i - 1] + delta_encoded[i])

# Display Results
print("Original Signal:", original_signal)
print("Delta Encoded:", delta_encoded)
print("Decoded Signal:", decoded_signal)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Signal Parameters
f = 10  
fs = 15  
duration = 1 
# Time vectors
t_cont = np.linspace(0, duration, 1000)  
t_samp = np.arange(0, duration, 1/fs)    

# Sine wave signals
x_cont = np.sin(2 * np.pi * f * t_cont)      
x_samp = np.sin(2 * np.pi * f * t_samp)    

plt.figure(figsize=(10, 5))
plt.plot(t_cont, x_cont, label='Original 10 Hz Sine Wave', linewidth=1)
plt.stem(t_samp, x_samp, linefmt='r-', markerfmt='ro', basefmt=' ', label='Sampled at 15 Hz')
plt.title('Aliasing: Sampling 10 Hz Signal at 15 Hz')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
