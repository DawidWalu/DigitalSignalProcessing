```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([6, 2, 4, 3, 4, 5, 0, 0, 0, 0])

# Function to create IDFT matrix
def create_idft_matrix(N):
    n = np.arange(N)
    k = n.reshape((N, 1))
    W = np.exp(2j * np.pi * n * k / N) / N
    return W

# Function to perform IDFT using matrix multiplication
def idft_matrix(X):
    N = len(X)
    W = create_idft_matrix(N)
    x = np.dot(W, X)
    return x, W

# Compute IDFT for N=10
x, W = idft_matrix(X)

# Display matrices and plot
print("IDFT Matrix W for N=10:")
print(np.round(W, 3))  

print("\nFrequency domain vector X:")
print(X)

print("\nTime domain signal x (real part):")
print(np.round(np.real(x), 4))  

plt.figure(figsize=(10, 5))
plt.stem(np.real(x))
plt.title('Synthesized Time Domain Signal (N=10)')
plt.xlabel('Sample index')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
def create_K_matrix(N):
    n = np.arange(N)
    k = n.reshape((N, 1))
    K = n * k
    return K

K_10 = create_K_matrix(10)
print("\nMatrix K for N=10:")
print(K_10)
```
